# Named Entity Recognition

We can use spaCy to process natural languages, identify part of speech of each tokens, and determine the named entity recognition of some tokens

```python
import spacy
from spacy import displacy
import pandas as pd
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv('./data/Articles.csv', sep=",", encoding= 'unicode_escape')
```

```python
def clean(doc):
    final_token = [token.lemma_ for token in doc]
    return " ".join(final_token)
```

```python
df['Article'] = df['Article'].apply(lambda x : clean(nlp(x)))
```

```python
doc = []
for i in range(50):
    doc += [nlp(df['Article'][i])]
```

```python
displacy.render(nlp(clean(doc[30])), style="ent")
```
